**Checkpoint: Programmation Orientée Objet (POO)**

**Objectif**

L'objectif est de créer une classe Python nommée "Account" avec des méthodes pour déposer, retirer et vérifier les soldes, encapsulée dans un script appelé "bank_account.py" pour simuler les opérations de base d'un compte bancaire.

```python
class Account:
  def __init__(self, numcompte, balance, nomclient):
    self.owner = numcompte
    self.balance = balance
    self.nomclient = nomclient
  def deposit(self, montantdepot):
    self.balance += montantdepot
    return self.balance
  def withdraw(self, montantdebit):
       if self.balance >= montantdebit:
           self.balance -= montantdebit
       else :
           print("Opération impossible: Solde inssuffisant")
       return self.balance
  def check_balance(self):
    return self.balance
```

```python
mon_compte = Account("12345", 1000000, "Essan Julien")
```

```python
# Depot
print(mon_compte.deposit(500000))

# retrait
print(mon_compte.withdraw(2200000))

# Vérification du solde
print(mon_compte.check_balance())
```
