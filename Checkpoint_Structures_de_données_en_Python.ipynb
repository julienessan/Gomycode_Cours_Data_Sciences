**Question 1  : **
Écrivez un programme Python qui multiplie tous les éléments d'une liste.

Esample list= [2, 3, 6]

Résultat = 36

```python
ma_list = [2, 3, 6]
product = 1
for element in ma_list :
  product *= element
print(product)
```

**Question 2**

Écrivez un programme Python pour obtenir une liste, triée par ordre croissant du dernier élément de chaque tuple, à partir d'une liste donnée de tuples non vides.

Liste type : [(2, 5), (1, 2), (4, 4), (2, 3), (2, 1)]

Résultat attendu : [(2, 1), (1, 2), (2, 3), (4, 4), (2, 5)]

Histoire : Vous pouvez utiliser la fonction sort.

```python
liste_type = [(2, 5), (1, 2), (4, 4), (2, 3), (2, 1)]
liste_type.sort(key= lambda x: x[-1])
print(liste_type)
```

**Question 3 **

Écrivez un programme Python qui combine deux dictionnaires en ajoutant des valeurs pour les clés communes.

d1 = {'a' : 100, 'b' : 200, 'c':300}

d2 = {'a' : 300, 'b' : 200, 'd':400}

Résultat attendu: {'a' : 400, 'b' : 400, 'd' : 400, 'c' : 300}

```python
d1 = {'a' : 100, 'b' : 200, 'c':300}
d2 = {'a' : 300, 'b' : 200, 'd':400}
dic_combine = d1.copy()
for key, value in d2.items():
  if key in dic_combine:
    dic_combine[key] += value
  else:
    dic_combine[key] = value
print(dic_combine)
```

**Question 4**

Avec un nombre entier n donné, écrivez un programme pour générer un dictionnaire qui contient (i, i*i) de sorte que soit un nombre entier entre 1 et n (les deux inclus). Le programme doit ensuite imprimer le dictionnaire. Supposons que l'entrée suivante soit fournie au programme : 8. La sortie devrait alors être : {1 : 1, 2 : 4, 3 : 9, 4 : 16, 5 : 25, 6 : 36, 7 : 49, 8 : 64}

```python
nb_entier = input("Entrer un nombre entier qui servira à générer un dictionnaire")
mon_dic = {}
for i in range(1, int(nb_entier)+1):
  mon_dic[i] = i*i
print(mon_dic)
```

**Question 5**

Écrivez un programme pour trier un tuple par son élément flottant.

Par exemple : list= [('item1', '12.20'), ('item2', '15.10'), ('item3', '24.5')]

Résultat attendu: [('item3', '24.5'), ('item2', '15.10'), ('item1', '12.20')]

```python
ma_list_float = [('item1', '12.20'), ('item2', '15.10'), ('item3', '24.5')]
ma_list_float.sort(key= lambda x: float(x[-1]), reverse=True)
print(ma_list_float)
```

Écrivez un programme Python pour créer un ensemble.

Exemples : {0, 1, 2, 3, 4}

```python
nb_entier = input("Entrer un nombre entier qui servira à créer un ensemble")
mon_ensemble = set()
for i in range(int(nb_entier)+1):
  mon_ensemble.add(i)
print(mon_ensemble)
```
