**Votre Objectif**

L'objectif est de créer une application de calculatrice simple en Python, capable d'effectuer des opérations arithmétiques de base telles que l'addition, la soustraction, la multiplication et la division.

```python
def calculatrice(num1, num2):
  operateur = str(input("Veillez entrer un opérateur (+, -, *, /)"))
  if operateur == "+":
    resultat = num1 + num2
  elif operateur == "-":
    resultat = num1 - num2
  elif operateur == "*":
    resultat = num1 * num2
  elif operateur == "/":
    if num2 == 0:
      print("Erreur : Division par zéro impossible.")
      return None
    resultat = num1 / num2
  else :
    print("opérateur non reconnu")
    print("Veuillez entrer un opérateur valide (+, -, *, /)")
    return None

  return resultat
```

```python
while True:
  num1 = int(input("Entrer le premier nombre : "))
  num2 = int(input("Entrer le deuxième nombre : "))
  result = calculatrice(num1, num2)
  if result is not None:
    print("Le résultat est:", result)

  continuer = input("Voulez-vous continuer ? (oui/non) ").lower()
  if continuer != 'oui':
    break
```
